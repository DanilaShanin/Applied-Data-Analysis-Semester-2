1. Напишите промт вида:  
   *«Напиши короткий обзор книги, не используя прилагательные “хороший” и “плохой”, и не упоминая никакие бренды.»*

```python

# Определение ограничений
restrictions = ["хороший", "плохой", "Apple", "Samsung"]

# Промт
prompt = f"""Книга предлагает увлекательное путешествие по миру сложных эмоций и интригующих событий. Автор мастерски создает атмосферу, погружая читателя в детали сюжета и заставляя задуматься о глубинных вопросах человеческой природы. Стиль написания отличается ясностью и динамичностью, позволяя легко следить за развитием истории. """


response = pipe(prompt)

# Проверка соблюдения ограничений
def check_restrictions(text, restrictions):
    for word in restrictions:
        if word.lower() in text.lower():
            return False
    return True

if check_restrictions(response[0]["generated_text"], restrictions):
    print("Ограничения соблюдены.")
else:
    print("Обнаружены запрещённые слова.")
```

Для доходов меньше или равных 10 000 рублей налог составляет 0%.
Для доходов свыше 10 000 рублей применяется ставка 13%.

```python
def calculate_tax(income):
    """
    Эта функция принимает доход в виде аргумента и возвращает сумму налога.

    Аргументы:
    income (float): доход, с которого будет рассчитан налог.

    Возвращаемое значение:
    float: сумма налога.
    """
    # Проверяем, не является ли доход отрицательным числом
    if income < 0:
        raise ValueError("Доход не может быть отрицательным.")

    # Если доход меньше или равен 10 000 рублей, налог равен нулю
    if income <= 10000:
        return 0
    else:
        # Рассчитываем налоговую базу (доход минус необлагаемый минимум)
        taxable_income = income - 10000
        # Вычисляем сумму налога по ставке 13%
        tax_amount = taxable_income * 0.13
        # Округляем сумму налога до двух знаков после запятой
        return round(tax_amount, 2)

income = 15000  # Доход
tax = calculate_tax(income)
print(f"Сумма налога для дохода {income} рублей составляет {tax} рублей.")
```

```python
import torch
from transformers import AutoModelForCausalLM, AutoTokenizer, pipeline

torch.random.manual_seed(0)
model = AutoModelForCausalLM.from_pretrained(
    "microsoft/Phi-3-mini-4k-instruct",
    device_map="cuda",
    torch_dtype="auto",
    trust_remote_code=True,
)

tokenizer = AutoTokenizer.from_pretrained("microsoft/Phi-3-mini-4k-instruct")
```

```python

def generate_summary(text):

    prompt = f"Пожалуйста, сделайте краткое резюме данного текста: {text}"


    generation_args = {
        "max_new_tokens": 500,
        "temperature": 1.0,
        "top_p": 0.6,
        "do_sample": True,
    }



    # Используем pipeline для генерации текста
    pipe = pipeline(
        "text-generation",
        model=model,
        tokenizer=tokenizer,
    )


    result = pipe(prompt, **generation_args)[0]["generated_text"]

    # Убираем цифры и запрещенные слова
    result = result.replace('очень', '').replace('крайне', '')
    result = ''.join([i for i in result if not i.isdigit()])

    return result.strip()

# Пример текста
test_text = """ В последние годы наблюдается значительный рост интереса к технологиям искусственного интеллекта. С каждым годом увеличивается количество исследований в этой области, и многие компании начинают активно внедрять ИИ-решения в свою работу. Эксперты прогнозируют, что в ближайшие пять лет инвестиции в разработку и внедрение ИИ возрастут в несколько раз. Это связано с тем, что технологии становятся всё более доступными и эффективными. Одним из ключевых направлений развития является автоматизация процессов, позволяющая компаниям значительно повысить производительность. """

summary = generate_summary(test_text)


print(f"Краткое резюме:\n{summary}")
```
